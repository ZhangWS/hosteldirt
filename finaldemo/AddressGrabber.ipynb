```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import re
import datetime
import ftfy
```

```python
#read in city CSV to get list of links
#Please note that though the encoding is nominally UTF-8, 
#CSV really doesn't like it and outputs mojibake instead of proper accents

hostellinks = pd.read_csv('hostellinks.csv', header=0, encoding="utf-8")
#hostellinks.to_sql('hostel_list', engine, if_exists = 'replace')
```

```python
import time
now = time.strftime("%Y%m%d-%H%M%S")
csvname = 'hosteladdress'+now+'.csv'
```

```python
#hostelworld displays 10 reviews to a page. Now we start scraping and processing raw data
def hosteldujour(index):
    i = hostellinks['hostelid'][index]
    u = hostellinks['url'][index]
    return i,u
```

```python
with open(csvname, 'w', newline='', encoding="UTF-8") as f:
    writer = csv.writer(f)
    writer.writerow(['hostelid','address'])
    
    for ind in hostellinks.index:
        hostelid, hostelurl = hosteldujour(ind)
        response = requests.get(hostelurl)
        address_soup = BeautifulSoup(response.text, 'html.parser')
        
        addr = address_soup.find_all(lambda tag: tag.name == 'property-map')
        if not addr:
            break
        else: 
            temp = str(addr).split("address")[1].split(";")
            if len(temp) > 2:
                temp = temp[2].split("&")[0]
            elif len(temp) > 1:
                temp = temp[1].split("\"")[0]
            else:
                temp = temp[0][3:-3]
            print(temp)
            writer.writerow([hostelid, temp])
```
